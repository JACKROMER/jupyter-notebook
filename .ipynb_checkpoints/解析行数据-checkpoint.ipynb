```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

    
# timearrarr
# 18:00:05 18:00:05 18:00:05 18:00:05  18:00:06
def drawLine(xarr,yarr,c,la):
    # 
    plt.rcParams['font.sans-serif'] = [''] 
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111)
    ax.set_title(la + " - analysis", fontsize=15)
    xp = []
    yp = []
    count = []
    for index in range(len(xarr) -1):
        if(xarr[index].strip() == xarr[index + 1].strip()): # 
           count.append(yarr[index]) # 
        else: #  
            count.append(yarr[index])
            thenums = str(xarr[index]).replace(":","").replace(",","") 
            xp.append(thenums) 
            avr = float(np.mean(count))
            yp.append(avr)
            # 
            count= []
    # 
    for i in range(len(xp)):
        print(xp[i], "-", yp[i]) 
    cc = float(np.mean(yp))
#     xx = []
#     for i in range(len(yp)):
#         xx.append(cc)
#     ax.set_yxim(xp[0], xp[-1]) 
    plt.xticks(rotation=90)
    plt.plot(xp, yp, marker='o', color=c, label='')
#     ax.plot(xp, xx, ls='--', color=c, label='')
    ax.hlines(np.mean(yp), xp[0], xp[-1:], linestyles='-.', colors='red')
    ax.text(2, np.mean(yp) * 1.2,'average:' + str(round(np.mean(cc),6)), color='red', fontsize=15)
    plt.show()

def drawAllLine(xarr,yarr,c,la):
    plt.rcParams['font.sans-serif'] = [''] 
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111)
    ax.set_title(la + " - analysis", fontsize=15)
    plt.axhline(y=np.mean(yarr),color="red") 
#     ax.hlines(np.mean(yarr), xarr[0], xarr[-1:], linestyles='-.', colors='red') 
    ax.text(len(xarr) / 2, np.mean(yarr) * 1.2,'average:' + str(round(np.mean(yarr),6)), color='red', fontsize=15) 
    plt.xticks(rotation=90)
    plt.plot(xarr, yarr, marker='o', color=c, label='')
    plt.show()
    
with open('log_Detection.txt', 'r') as f:
    timeArr = []
    pnarr = []
    lenarr=[]
    wtharr=[]
    hgtarr=[]
    index = 0;
    for line in f:
        if('overlay frame' in line):
            index += 1
            if(len(timeArr) > 0):
                print("find some data should draw image")
                print('pn average is :'  + str(np.mean(pnarr)))
                print('pn fangcha is :'  + str(np.var(pnarr)))
                print('pn bzcha   is :'  + str(np.var(pnarr)))
                print("---------------------------------------") 
                print("find some data should draw image")
                print('pn average is :'  + str(np.mean(lenarr)))
                print('pn fangcha is :'  + str(np.var(lenarr)))
                print('pn bzcha   is :'  + str(np.var(lenarr)))
                print("---------------------------------------") 
                print("find some data should draw image")
                print('pn average is :'  + str(np.mean(wtharr)))
                print('pn fangcha is :'  + str(np.var(wtharr)))
                print('pn bzcha   is :'  + str(np.var(wtharr)))
                print("---------------------------------------") 
                print("find some data should draw image")
                print('pn average is :'  + str(np.mean(hgtarr)))
                print('pn fangcha is :'  + str(np.var(hgtarr)))
                print('pn bzcha   is :'  + str(np.var(hgtarr)))
 
                drawLine(timeArr, pnarr, 'b', 'pn')
                drawLine(timeArr, lenarr,'g', 'len') 
                drawLine(timeArr, wtharr,'r', 'wth')
                drawLine(timeArr, hgtarr,'c', 'hgt')
                timeArr = []
                pnarr= []
                lenarr= []
                wtharr= [] 
                hgtarr= []
        if('2020' in line):
            arr = line.split(' ')
            timeArr.append(arr[1])
            index += 1 
            for item in arr:
                if("pn:" in item):
                    subarr = item.split(':')
                    pnarr.append(float(subarr[-1]))
                if("len:" in item):
                    subarr = item.split(':')
                    lenarr.append(float(subarr[-1]))
                if("wth:" in item):
                    subarr = item.split(':')
                    wtharr.append(float(subarr[-1]))
                if("hgt:" in item):
                    subarr = item.split(':')
                    hgtarr.append(float(subarr[-1]))
                    
drawAllLine(timeArr, pnarr, 'b', 'pn')           
drawAllLine(timeArr, lenarr, 'g', 'len')
drawAllLine(timeArr, wtharr, 'r', 'wth')   
drawAllLine(timeArr, hgtarr, 'c', 'hgt') 
```
